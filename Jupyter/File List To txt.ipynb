# 파일 이름 변경

```python
for i, filename in enumerate(os.listdir(path)):
    if filename[-3:]=='jpg':
        os.rename(path+filename, path+"negative_%04d"%i+'.jpg')
```

# 빈 .txt 파일 만들기

```python
for i, filename in enumerate(os.listdir(path)):
    print(filename)
    ff = open(path+filename[:-3]+'txt', 'w')
    f.close()
```
